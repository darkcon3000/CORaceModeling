```python
def electionCleaner(fileName, output):
    myFile = open(fileName,"r+") # open the file in read mode
    outputFile = open(output,"w")
    csvReader = csv.reader(myFile, delimiter=",",quotechar='"') #setup the csv reader with default settings
    csvWriter = csv.writer(outputFile, delimiter=",",quotechar='"',lineterminator = '\n')#, delimiter=",",quotechar='"')
    csvWriter.writerow(next(csvReader))
    for row in csvReader:
        if(row[5]=="Governor"):
            csvWriter.writerow(row)
```

```python
electionCleaner('2018GEPrecinctLevelResults.csv','coGovElection.csv')
```

```python
def countySorterElection(fileName, output):
    myFile = open(fileName,"r+") # open the file in read mode
    outputFile = open(output,"w")
    csvReader = csv.reader(myFile, delimiter=",",quotechar='"') #setup the csv reader with default settings
    csvWriter = csv.writer(outputFile, delimiter=",",quotechar='"',lineterminator = '\n')#, delimiter=",",quotechar='"')
    csvWriter.writerow(['County','Democratic Party','Republican Party','Unity Party of Colorado','Libertarian Party'])
    header = next(csvReader)
    first = True
    dem = 0
    gop = 0
    thirdParty = 0
    lib = 0
    for row in csvReader:
        #print(row)
        if (first):
            first = False
            county = row[1]
            if (row[3] == "Democratic Party"):
                dem += int(row[4])
            elif (row[3] == "Republican Party"):
                gop += int(row[4])
            elif (row[3] == "Unity Party of Colorado"):
                thirdParty += int(row[4])
            elif (row[3] == "Libertarian Party"):
                lib += int(row[4])
        else:
            if (row[1] == county):
                if (row[3] == "Democratic Party"):
                    dem += int(row[4])
                elif (row[3] == "Republican Party"):
                    gop += int(row[4])
                elif (row[3] == "Unity Party of Colorado"):
                    thirdParty += int(row[4])
                elif (row[3] == "Libertarian Party"):
                    lib += int(row[4])
            else:
                csvWriter.writerow([county,str(dem),str(gop),str(thirdParty),str(lib)])
                dem = 0
                gop = 0
                thirdParty = 0
                lib = 0
                county = row[1]
                if (row[3] == "Democratic Party"):
                    dem += int(row[4])
                elif (row[3] == "Republican Party"):
                    gop += int(row[4])
                elif (row[3] == "Unity Party of Colorado"):
                    thirdParty += int(row[4])
                elif (row[3] == "Libertarian Party"):
                    lib += int(row[4])
    myFile.close()
```

```python
countySorterElection('coGovElection2.csv','coGovElectionFinal.csv')
```
