```python
def countySorter(fileName, output):
    myFile = open(fileName,"r+") # open the file in read mode
    outputFile = open(output,"w")
    csvReader = csv.reader(myFile, delimiter=",",quotechar='"') #setup the csv reader with default settings
    csvWriter = csv.writer(outputFile, delimiter=",",quotechar='"',lineterminator = '\n')#, delimiter=",",quotechar='"')
    csvWriter.writerow(['County','CFIPS','Race','Counts'])
    header = next(csvReader)
    first = True
    for row in csvReader:
        if first:
            first = False
            #county = row[0]
            csvWriter.writerow([row[1],row[2],row[4],row[5]])
        else:
            first = True
            #county = row[0]
            csvWriter.writerow([row[1],row[2],row[3],row[5]])
    myFile.close()
```

```python
def countySorterFinal(fileName, output):
    myFile = open(fileName,"r+") # open the file in read mode
    outputFile = open(output,"w")
    csvReader = csv.reader(myFile, delimiter=",",quotechar='"') #setup the csv reader with default settings
    csvWriter = csv.writer(outputFile, delimiter=",",quotechar='"',lineterminator = '\n')#, delimiter=",",quotechar='"')
    csvWriter.writerow(['County','CFIPS','White','Black','Asian/Pacific Islander','Hispanic Origin','American Indian'])
    header = next(csvReader)
    first = True
    asian = 0
    white = 0
    black = 0
    native = 0
    hispanic = 0
    for row in csvReader:
        if (first):
            first = False
            county = row[0]
            hispanic += float(row[3])
        else:
            if (row[0] == county):
                if (row[2] == "Hispanic Origin"):
                    hispanic += float(row[3])
                elif (row[2] == "American Indian"):
                    native += float(row[3])
                elif (row[2] == "Asian/Pacific Islander"):
                    asian += float(row[3])
                elif (row[2] == "Black"):
                    black += float(row[3])
                elif (row[2] == "White"):
                    white += float(row[3])
            else:
                csvWriter.writerow([county,row[1],str(white),str(black),str(asian),str(hispanic),str(native)])
                asian = 0
                white = 0
                black = 0
                native = 0
                hispanic = 0
                county = row[0]
                if (row[2] == "Hispanic Origin"):
                    hispanic += float(row[3])
                elif (row[2] == "American Indian"):
                    native += float(row[3])
                elif (row[2] == "Asian/Pacific Islander"):
                    asian += float(row[3])
                elif (row[2] == "Black"):
                    black += float(row[3])
                elif (row[2] == "White"):
                    white += float(row[3])
    myFile.close()
```
